```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
Physics_std = 10
Physics_mean = 80

Physics = (np.round(np.random.randn(500) * Physics_std + Physics_mean))
```

```python
Maths_std = 20
Maths_mean = 40

Maths = (np.round(np.random.randn(500) * Maths_std + Maths_mean))
```

```python
plt.hist(Physics,color= 'blue')
plt.hist(Maths,color='orange',alpha=0.8)
plt.show()
```
